```python
import nibabel as nib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from skimage import measure

def plot_3d(nifti_image, threshold=-600):
    nifti_data = nifti_image.get_fdata()
    p = nifti_data.transpose(2, 1, 0) 
    verts, faces, normals, values = measure.marching_cubes(p, threshold)
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    mesh = Poly3DCollection(verts[faces], alpha=0.1)
    face_color = [0.5, 0.5, 1]
    mesh.set_facecolor(face_color)
    ax.add_collection3d(mesh)
    ax.set_xlim(0, p.shape[0])
    ax.set_ylim(0, p.shape[1])
    ax.set_zlim(0, p.shape[2])
    plt.show()
plot_3d(sample_data)
```

Plot in X,Y,Z plane

```python
import nibabel as nib
import matplotlib.pyplot as plt

def visualize_nifti_slices(nifti_image):
    nifti_data = nifti_image.get_fdata()
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(nifti_data[nifti_data.shape[0] // 2, :, :], cmap='gray')
    axes[0].set_title('Sagittal Plane (X)')
    axes[0].axis('off')

    axes[1].imshow(nifti_data[:, nifti_data.shape[1] // 2, :], cmap='gray')
    axes[1].set_title('Coronal Plane (Y)')
    axes[1].axis('off')

    axes[2].imshow(nifti_data[:, :, nifti_data.shape[2] // 2], cmap='gray')
    axes[2].set_title('Axial Plane (Z)')
    axes[2].axis('off')
    plt.show()
visualize_nifti_slices(sample_data)
```
